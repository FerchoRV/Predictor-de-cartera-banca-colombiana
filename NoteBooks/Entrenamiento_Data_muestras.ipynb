```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.preprocessing import MinMaxScaler,  StandardScaler, LabelEncoder
import joblib
```

```python
ruta_data = 'D:/Marlon Volveras - IA finan/Predictor de cartera banca Colombiana PYMES/Datos/Muestras_Banco_TCredito/cartera_Banagrario_MICRO_MENOR.csv'

df= pd.read_csv(ruta_data)
df.head()
```

```python
# Calcula el índice de corte basado en el porcentaje deseado
indice_corte = int(len(df) * 0.8)  # 80% de los datos

# Divide el DataFrame en dos partes
df_train = df.iloc[:indice_corte]  # 80% de los datos para entrenamiento
df_test = df.iloc[indice_corte:]  # 20% de los datos para pruebas

# Verifica las longitudes de los DataFrames de entrenamiento y prueba
print("Longitud del conjunto de entrenamiento:", len(df_train))
print("Longitud del conjunto de pruebas:", len(df_test))
```

# Definir valirables 'X' y 'y' de entrenamiento y prueba

# Dividir datos en conjutno de entrenamiento y prueba

Como son valores de seires de tiempo al momento de dividir se dejan el ultimo 20% de todos los datos y se corta por medio del index de los datos



```python
X_train = df_train.drop('(1) Saldo de la cartera a la fecha de corte del reporte', axis=1)
X_test = df_test.drop('(1) Saldo de la cartera a la fecha de corte del reporte', axis=1)

y_train= df_train['(1) Saldo de la cartera a la fecha de corte del reporte']
y_test= df_test['(1) Saldo de la cartera a la fecha de corte del reporte']

print('Conjunto de datos de entrada de entrenamiento', X_train.shape)
print('Conjunto de datos de salida de entrenamiento', y_train.shape)

print('Conjunto de datos de entrada de prueba', X_test.shape)
print('Conjunto de datos de salida de prueba', y_test.shape)
```

```python
# definir modelo de red LSTM

model = keras.Sequential() #Linea 1

model.add(layers.LSTM(150, return_sequences=True, input_shape=(X_train.shape[1], 1)))

model.add(layers.LSTM(150, return_sequences=False)) #Linea 3

model.add(layers.Dense(30)) #Linea 4

model.add(layers.Dense(1)) #Linea 5

model.summary() #Linea 6
```

```python
# compilar modelo 

model.compile(optimizer='adam', loss='mean_squared_error') #Linea 1

# entrenar modelo 
     
model.fit(X_train, y_train, batch_size= 1, epochs=10) #Linea 2
```

```python
y_pred = model.predict(X_test)

# Error absoluto medio (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)

# Raíz del error cuadrático medio (RMSE)
rmse = np.sqrt(mse)

# Coeficiente de determinación (R2 Score)
r2 = r2_score(y_test, y_pred)

print('# Error absoluto medio (MAE)', mae)
print('# Error cuadrático medio (MSE)', mse)
print('# Raíz del error cuadrático medio (RMSE)', rmse)
print('# Coeficiente de determinación (R2 Score)', r2)
```
