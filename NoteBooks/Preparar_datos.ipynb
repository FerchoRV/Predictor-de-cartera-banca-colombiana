Este estudio se basa en el comportamiento de cartera de los creditos entregamos a PYMES en colombia por esto filtramos el conjutno de datos por medio de la columna DESC_RENGLON a solo cr4ditos que estas entidades pueden adquirir.

```python
# seleccionar datos de interes filtro tipo de credito
categorias_des_renglon = df['DESC_RENGLON'].unique().tolist()

categorias_des_renglon
```

```python
tipo_credit_in = ['PYMES','CORPORATIVO','EMPRESARIAL','FACTORING','MICROEMPRESA','MICRO MAYOR TOTAL','MICRO MENOR TOTAL','CRÉDITO ROTATIVO TOTAL']

muestra_interes = muestra = df[df['DESC_RENGLON'].isin(tipo_credit_in)]

muestra_interes.info() 
```

```python
# processar columna de fecha

muestra_interes['FECHA_CORTE'] = pd.to_datetime(muestra_interes['FECHA_CORTE'])
```

```python
muestra_interes.head()
```

# Se crear cartateristicas extrar a la muestra por medio de la extraccion de infromacion de la columna de fecha de corte y se combina las columnas enditdad y descripcion de renglon para poder ordenar los datos por fecha sin perder la relacion de estas y no sea neecsario subdividir el data framen por entidad y tipo de credito

```python
muestra_interes['año'] = muestra_interes['FECHA_CORTE'].dt.year
muestra_interes['mes'] = muestra_interes['FECHA_CORTE'].dt.month
muestra_interes['dia'] = muestra_interes['FECHA_CORTE'].dt.day
```

```python
muestra_interes['banco_tipo_credito'] = muestra_interes['NOMBREENTIDAD'] + '_' + muestra_interes['DESC_RENGLON']
```

```python
muestra_interes
```

```python
relaciones_banco_credi = muestra_interes['banco_tipo_credito'].unique().tolist()

len(relaciones_banco_credi)
```

```python
muestra_interes = muestra_interes.sort_values(by='FECHA_CORTE')

muestra_interes.head()
```

# Eliminar columnas redundantes

Se elimina toda columna que este duplicada por las nuevas caracteristica creadas 

```python
colum_redun = ['TIPO_ENTIDAD','CODIGO_ENTIDAD','NOMBREENTIDAD','FECHA_CORTE','UNICAP','DESCRIP_UC','RENGLON','DESC_RENGLON']

df_final = muestra_interes.drop(columns=colum_redun)
```
